```python
import pygame
import random
pygame.init()
 

win_w, win_h = 288, 512
window = pygame.display.set_mode((win_w, win_h))


background = pygame.image.load('/Users/rnda/Desktop/CODING_PROJECTS/NEAT_FLAPPY_B/imgs/bg.png')
Basee = pygame.image.load('/Users/rnda/Desktop/CODING_PROJECTS/NEAT_FLAPPY_B/imgs/base.png')
bird1 = pygame.image.load('/Users/rnda/Desktop/CODING_PROJECTS/NEAT_FLAPPY_B/imgs/bird1.png') #CHANGE
bird2 = pygame.image.load('/Users/rnda/Desktop/CODING_PROJECTS/NEAT_FLAPPY_B/imgs/bird2.png')
bird3 = pygame.image.load('/Users/rnda/Desktop/CODING_PROJECTS/NEAT_FLAPPY_B/imgs/bird3.png')
pipe_img = pygame.image.load('/Users/rnda/Desktop/CODING_PROJECTS/NEAT_FLAPPY_B/imgs/pipe.png')
You_lost_font = pygame.font.SysFont("comicsans",20)
score = 0


#use rect to get right proportions. fix gap 

class Pipe:  

    GAP = 100 
    VEL = 1
    bottom_pipe_img = pipe_img
    top_pipe_img = pygame.transform.rotate(pipe_img,180)

    def __init__(self, bottom_x1 = win_w/3, bottom_x2 = win_w/2 + win_w/5 , bottom_y1 = random.randint(win_h/2 +20, win_h-220),bottom_y2 = random.randint(win_h/2 +20, win_h-200)):
 
        self.bottom_x1 = bottom_x1
        self.top_x1 = self.bottom_x1


        self.bottom_x2 = bottom_x2
        self.top_x2 = self.bottom_x2


        self.bottom_y1 = bottom_y1
        self.top_y1 = self.bottom_y1 - self.GAP - self.top_pipe_img.get_height()


        self.bottom_y2 = bottom_y2
        self.top_y2 = self.bottom_y2 - self.GAP - self.top_pipe_img.get_height()

        self.score = 0

    def move(self):

        self.bottom_x1 -= self.VEL
        self.bottom_x2 -= self.VEL
        self.top_x2 -= self.VEL
        self.top_x1 -= self.VEL
      

        if self.bottom_x1  < 38 - bird1.get_width() :
            self.bottom_x1 = win_w/2 + win_w/4
            self.top_x1 = self.bottom_x1
            self.score +=1

            self.bottom_y1 = random.randint(win_h/2, win_h-200)
            self.top_y1 = self.bottom_y1 - self.GAP - self.top_pipe_img.get_height()

        if self.bottom_x2 < 38 - bird1.get_width():
            self.bottom_x2 = win_w/2 + win_w/4
            self.top_x2 = self.bottom_x2
            self.score +=1

            self.bottom_y2 = random.randint(win_h/2-20, win_h-230)
            self.top_y2 = self.bottom_y2 - self.GAP - self.top_pipe_img.get_height() 

    def draw(self):
        window.blit(self.bottom_pipe_img,(self.bottom_x1,self.bottom_y1))
        window.blit(self.bottom_pipe_img,(self.bottom_x2,self.bottom_y2))
        window.blit(self.top_pipe_img,(self.top_x1,self.top_y1))
        window.blit(self.top_pipe_img,(self.top_x2,self.top_y2))



    def reset(self):
        self.bottom_x1 = win_w/3
        self.top_x1 = self.bottom_x1
        
        self.bottom_x2 = win_w/2 + win_w/5
        self.top_x2 = self.bottom_x2

        self.bottom_y1 = random.randint(win_h/2 +20, win_h-220)
        self.top_y1 = self.bottom_y1 - self.GAP - self.top_pipe_img.get_height()

        self.bottom_y2 = bottom_y2 = random.randint(win_h/2 +20, win_h-200)
        self.top_y2 = self.bottom_y2 - self.GAP - self.top_pipe_img.get_height()

            
        

class Background:
    VEL = 5

    def __init__(self,y):
        self.y = y
        self.img = background
        self.x1 = 0
        self.width = self.img.get_width()
        self.x2 = self.img.get_width()


    def move(self):
        self.x1 -= self.VEL
        self.x2 -= self.VEL

        if self.x1 + self.width < 0:
            self.x1 = self.x2 + self.width

        if self.x2 + self.width < 0:
            self.x2 = self.x1 + self.width

    def draw(self):
        window.blit(self.img,(self.x1,self.y))
        window.blit(self.img,(self.x2,self.y))



class Base:
    VEL = 5

    def __init__(self,y):
        self.y = y
        self.img = Basee
        self.x1 = 0
        self.width = self.img.get_width()
        self.x2 = self.img.get_width()


    def move(self):
        self.x1 -= self.VEL
        self.x2 -= self.VEL

        if self.x1 + self.width < 0:
            self.x1 = self.x2 + self.width

        if self.x2 + self.width < 0:
            self.x2 = self.x1 + self.width

    def draw(self):
        window.blit(self.img,(self.x1,self.y))
        window.blit(self.img,(self.x2,self.y))


class Bird:
    jump_vel = -10.5
    bird_imgs = [bird1,bird2,bird3]
    jump_count = 0
    flap_count = 0
    max_tilt = og_max_tilt = 25
    score = 0


    def __init__(self,y):
        self.x = 33
        self.y = self.og_y = y
        self.bird_img_count = 0
        self.img = self.bird_imgs[self.bird_img_count]
        self.tilt = 0

    def jump(self):

        #create hill when pressing space

        #track when we last jump. when jump, restart jump_count and displacement 
        keys = pygame.key.get_pressed()
        if keys[pygame.K_SPACE]:
            self.jump_count = 0
            self.jump_count += 1
            self.tilt = self.max_tilt
            displacement = self.jump_count * self.jump_vel +  0.5*(3)* self.jump_count **2
            self.y += displacement
        
        else:
            self.jump_count += 1
            self.tilt -= 2
            displacement = self.jump_count * 0.1
            self.y += displacement
        
        

        #if key down space happens once how does displacement cause full jump shape?


    def flap(self):

        self.flap_count += 1

        if self.flap_count == 5:
            self.bird_img_count += 1
            self.img = self.bird_imgs[self.bird_img_count]
    
        elif self.flap_count == 10:
            self.bird_img_count += 1
            self.img = self.bird_imgs[self.bird_img_count]

        elif self.flap_count == 15:
            self.bird_img_count -= 2
            self.flap_count -= 15
            self.img = self.bird_imgs[self.bird_img_count]


    def draw(self):

        self.img_r = pygame.transform.rotate(self.img,self.tilt)
        window.blit(self.img_r,(self.x,self.y))


    def get_mask(self):
        return pygame.mask.from_surface(self.img)


    def reset(self):
        self.x = 33
        self.y = self.og_y
        self.bird_img_count = 0
        self.img = self.bird_imgs[self.bird_img_count]
        self.tilt = 0



def pipe_bird_collision(pipe,bird):
    bird_mask = bird.get_mask()
    top_mask = pygame.mask.from_surface(pipe.top_pipe_img) #creates mask using top pipe
    bottom_mask = pygame.mask.from_surface(pipe.bottom_pipe_img) #creates mask using bottom pipe
    top_offset = (self.x - bird.x, self.top - round(bird.y))
    bottom_offset = (self.x - bird.x, self.bottom - round(bird.y)) 

    b_point = bird_mask.overlap(bottom_mask, bottom_offset) #checks if point of overlap between bird mask n bot pipe.
    t_point = bird_mask.overlap(top_mask,top_offset)  #will return none if no collision.

    if b_point or t_point: #checks if collisions returned true
        return True

    return False


def move_and_draw_images(back,base,pipe,bird):
    back.draw()
    back.move()
    base.move()
    pipe.draw()
    pipe.move()
    base.draw()
    bird.draw()
    bird.flap()


def main():
    run = True
    clock = pygame.time.Clock() 

    back = Background(0)
    base = Base(win_h - Basee.get_height())
    pipe = Pipe()
    bird = Bird(((win_h-Basee.get_height())//2 - bird1.get_height()//2))
    You_lost_label = You_lost_font.render(f"You lost, restarting in 3 sec", 1, (255,255,255))
    score_label = You_lost_font.render(f"score: {pipe.score}", 1, (255,0,0))

    while run:

        clock.tick(40)

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False
                break
        
        move_and_draw_images(back,base,pipe,bird)
        bird.jump()

        if pipe.bottom_x1 + pipe_img.get_width() >33 and pipe.bottom_x1 < 33 + bird.img.get_width():
            if bird.y + bird.img.get_height() > pipe.bottom_y1:
                pipe.reset()
                bird.reset()
                time = 5
                while time >= 0: 
                    window.blit(You_lost_label,(144-You_lost_label.get_width()/2,210-You_lost_label.get_height()/2))
                    window.blit(score_label,(144-You_lost_label.get_width()/2,240-You_lost_label.get_height()/2))
                    pygame.display.update()
                    time -= 1

           
                pygame.time.delay(3000)

        if pipe.bottom_x2 + pipe_img.get_width() >33 and pipe.bottom_x2 < 33 + bird.img.get_width():
            if bird.y + bird.img.get_height()> pipe.bottom_y2:
                pipe.reset()
                bird.reset()
                time = 5
                while time >= 0: 
                    window.blit(You_lost_label,(144-You_lost_label.get_width()/2,210-You_lost_label.get_height()/2))
                    window.blit(score_label,(144-You_lost_label.get_width()/2,240-You_lost_label.get_height()/2))
                    pygame.display.update()
                    time -= 1

                pygame.time.delay(3000)
                
        if pipe.top_x1 + pipe_img.get_width() >33 and pipe.top_x1 < 33 + bird.img.get_width():
            if bird.y < pipe.bottom_y1 - pipe.GAP-15:
                pipe.reset()
                bird.reset()
                time = 5
                while time >= 0: 
                    window.blit(You_lost_label,(144-You_lost_label.get_width()/2,210-You_lost_label.get_height()/2))
                    window.blit(score_label,(144-You_lost_label.get_width()/2,240-You_lost_label.get_height()/2))
                    pygame.display.update()
                    time -= 1

                pygame.time.delay(3000)

        if pipe.top_x2 + pipe_img.get_width() >33 and pipe.top_x2 < 33 + bird.img.get_width():
            if bird.y < pipe.bottom_y2 - pipe.GAP-15:
                pipe.reset()
                bird.reset()
                time = 5
                while time >= 0: 
                    window.blit(You_lost_label,(144-You_lost_label.get_width()/2,210-You_lost_label.get_height()/2))
                    window.blit(score_label,(144-You_lost_label.get_width()/2,240-You_lost_label.get_height()/2))
                    
                    pygame.display.update()
                    time -= 1

                pygame.time.delay(3000)
            
        if bird.y >= base.y-10:
            bird.reset()
            pipe.reset()
            time = 5
            while time >= 0: 
                window.blit(You_lost_label,(144-You_lost_label.get_width()/2,210-You_lost_label.get_height()/2))
                window.blit(score_label,(144-You_lost_label.get_width()/2,240-You_lost_label.get_height()/2))
                
                pygame.display.update()
                time -= 1

            pygame.time.delay(3000)
            
        pygame.display.update()

    pygame.display.quit()


main()



```
